### Q1.

**GET :- The GET method is used to request data from the server. When a GET request is made, the client sends a request to the server asking for a specific resource. The requested data is returned to the client in the form of a response. The data is usually appended to the URL as a query string. This means that the data is visible in the URL and can be bookmarked or shared. GET requests are used for reading data from the server and are considered safe and idempotent, meaning they can be executed multiple times without changing the result.**

**POST :- The POST method is used to submit data to the server. When a POST request is made, the client sends data to the server in the body of the request. The server processes the data and sends a response back to the client. POST requests are used for creating or updating data on the server and are not considered safe or idempotent because they can modify server data. The data submitted via POST is not visible in the URL and is typically used to submit form data or upload files.**

### Q2.

**In Flask, the request object is used to handle incoming HTTP requests sent to the server. When a client sends a request to the server, the request object allows the server to access information about the request such as the HTTP method (GET, POST, etc.), the URL, any headers that were sent with the request, and any data that was included in the request (such as form data or JSON).**

### Q3.

**The redirect() function in Flask is used to redirect the user's browser to a different URL. This is a common way to handle certain types of user actions, such as submitting a form or logging in, that require the application to show a different page after the action is completed.**
